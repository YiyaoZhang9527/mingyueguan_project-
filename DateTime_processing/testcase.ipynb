```python
week_table_path = expanduser('~/mingyueguan_project/中英文星期对照1.csv')
week_table = pd.read_csv(week_table_path)
week_table = week_table[pd.isna(week_table.中文)==False]
week_table = week_table.replace(np.nan,'')
week_table.insert(loc=1, column='中文2', value=["周日","周一","周二","周三","周四","周五","周六"])
week_table.insert(loc=1, column='数字', value=[0,1,2,3,4,5,6])
week_table.to_csv("中英文星期对照0.1.csv", index=False)
```

```python
def is_week(string, week_table):
    check_index = week_table == string
    vaild_week = check_index .to_numpy().sum()
    #m,n = check_index.shape
    #m_true,n_true = np.where(check_index.to_numpy()==True)
    if vaild_week:
        return week_table.数字[check_index.sum(axis=1) > 0].values[0]
     
is_week('周一', week_table)
```

```python
def is_month(string, month_table):
    check_index = month_table == string
    vaild_month = check_index .to_numpy().sum()
    if vaild_month:
        return month_table.数字[check_index.sum(axis=1) > 0].values[0]
    
is_month('March', month_table)    
```
